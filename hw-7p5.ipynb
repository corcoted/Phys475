```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
import numpy as np
```

```python
# Problem 1
psi1 = lambda x: np.cos(x*np.pi*0.5)
v1 = 0.
e1 = np.pi**2/8.
x1 = np.linspace(-1.,1.)
```

```python
plt.plot(x1,psi1(x1)+e1,label=u"ψ(x)")
plt.plot(x1,e1+0.*x1,label="E")
plt.plot([-1.001,-1.0,1.0,1.001],[3.,0.,0.,3.],'k',label="V(x)")
plt.xlabel("x")
plt.ylabel("ψ,energy")
plt.title("Problem 1")
plt.legend()
plt.savefig("hw7p5-1.png")
plt.show()
```

```python
# Problem 2
psi2 = lambda x: (1.+np.cos(np.pi*x))/np.sqrt(3)
v2 = lambda x: 0.5*x*x
e2 = 1.68
x2 = np.linspace(-1.,1.)
xx2 = np.linspace(-3.,3.)
```

```python
plt.plot(x2,psi2(x2)+e2,label=u"ψ(x)")
plt.plot(xx2,e2+0.*xx2,label="E")
plt.plot(xx2,v2(xx2),'k',label="V(x)")
plt.xlabel("x")
plt.ylabel("ψ,energy")
plt.title("Problem 2")
plt.legend()
plt.savefig("hw7p5-2.png")
plt.show()
```
